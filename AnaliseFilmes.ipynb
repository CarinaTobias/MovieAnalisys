### Dataset de filmes do IMDb

### O objetivo deste projeto é identificar a probabilidade de pessoas entre 18 e 30 anos assistirem filmes produzidos antes de 1980

## Importacao de bibliotecas

```python
import pandas as pd
import sys
```

## Carga dos dados

```python
# Carga do primeiro dataset
df1 = pd.read_csv("C:/Users/Carina/OneDrive/Notebook Dell/Tech/Python/movie/IMDb Ratings.csv")
df1.head()
```

```python
#Carga do segundo dataset
df2 = pd.read_csv("C:/Users/Carina/OneDrive/Notebook Dell/Tech/Python/movie/IMDbMovies.csv")
df2
```

## Seleção dos dados

```python
# Seleciona no primeiro dataset somente as colunas necessárias para este projeto e renomeia para nomes mais intuitivos

df3 = df1[['imdb_title_id', 'total_votes', 'allgenders_18age_votes', 'males_18age_votes', 'females_18age_votes']]
df3.head()
```

```python
# Seleciona no segundo dataset somente as colunas necessárias para este projeto e renomeia para nomes mais intuitivos

df4 = df2[['imdb_title_id', 'title', 'year', 'genre']]
df4.head()
```

## Junção dos datasets 

```python
# Junta os datasets em um único utilizando o ID do título do filme como chave 
df5 = pd.merge(df3, df4)
df5.head()
```

```python
# Analisa a estrutura de dados
df5.isna().sum()
```

```python
df5.dtypes
```

## Limpeza dos dados

```python
df5_remove = df5.loc[(df5['year'] == 'TV Movie 2019')]
df5_clean = df5.drop(df5_remove.index)   
```

```python
df5_clean['yearnum'] = df5_clean['year'].astype(int)
df5_clean.dtypes
```

## Agrupamento dos dados por período de produção do filme

```python
# Soma a quantidade de votos entre pessoas de 18 a 30 anos pelo período de produção do filme (antes de 1980 e após 1980)

df5_clean['yearGroup']=pd.cut(

   df5_clean['yearnum'],

   bins=[0, 1980, sys.maxsize],

   labels=['Antes1980', 'Apos1980'])

df5_clean
```

```python
df6 = df5_clean.groupby(["yearGroup"])["allgenders_18age_votes"].sum()
df6
```

## Calcula os totais para cálculo de probabilidade

```python
# Total de votos (espaço amostral total)
df5_clean["total_votes"].sum()
```

```python
# Total de votos das pessoas entre 18 e 30 anos (espaço amostral das pessoas entre 18 e 30 anos)
df5_clean["allgenders_18age_votes"].sum()
```

```python
# Probabilidade das pessoas entre 18 e 30 anos assistirem filmes produzidos antes de 1980 comparado ao total de votos
# (espaço amostral total)
probTotal = (df6[0] / df5_clean["total_votes"].sum())*100

print(probTotal)
```

```python
# Probabilidade das pessoas entre 18 e 30 anos assistirem filmes produzidos antes de 1980 comparado ao ano de produção
# (espaço amostral das pessoas entre 18 e 30 anos)
prob18 = (df6[0] / df5_clean["allgenders_18age_votes"].sum())*100
print(prob18)
```
